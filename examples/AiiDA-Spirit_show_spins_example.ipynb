# Example usage of the `show_spins` plotting function in aiida_spirit

If you want to execute this example you will need to do
```
verdi archive import path/to/aiida-spirit/examples/input_files/spirit_test_calculation.aiida
```
before continuing.

````python
from aiida import load_profile
_ = load_profile()
````

````python
# load an example SpiritCalculation
from aiida.orm import load_node
spirit_calc = load_node('73168e63-4f77-40a5-839a-1044ad17dfb1')
````

````python
from aiida_spirit.tools.plotting import init_spinview, show_spins

# initialize the area where the spins are shown
# this is only done once, then the show_spins function will put the spins into the widget
# multiple calls of show_spins will update the plot

init_spinview()
````

````python
# now populate the widget with the spins
# ATTENTION: wait for the init_spinview function to load fully
# otherwise the spins are not updated in the widget

show_spins(spirit_calc, scale_spins=0.8)
````
